```python
'''
Name: Lauren Fisher
Assignment: Project6 (Preliminary Nim Game)
Class: CS323 ML and Intelligent Systems

'''
import random

class nimGame():
    '''
    object containing the board, current board states, and allows for maniupulation of nim by players
    '''
    def __init__(self):
        """
        initializes list of board game state for game of nim

        Parameters
        ----------
        none

        Returns
        -------
        none

        """
        self.board = []
        self.board.append(['|'])
        self.board.append(['|', '|', '|'])
        self.board.append(['|', '|', '|', '|', '|'])
        self.board.append(['|', '|', '|', '|', '|', '|', '|'])
        

    def printWorld(self):
        """
        returns nicely printed board state list to be printed

        Parameters
        ----------
        none

        Returns
        -------
        none

        """
        for i in range(len(self.board)):
            print()
            print()
            print(i,end="")
            for j in range(len(self.board[i])):
                print(" " + self.board[i][j], end=" ")
                      


    def cp_move(self):
        """
        computer player randomly selects row to take from and number of tallies to take and updates board

        Parameters
        ----------
        none

        Returns
        -------
        self.is_over(): Bool of whether all rows are empty and game is over

        """
        valid_rows = []
        tallies_left = []
        for i in range(len(self.board)):
            if(map(len, self.board[i]) != 0):
                tallies_left.append(len(self.board[i]))
            if(len(self.board[i]) != 0):
                valid_rows.append(i)
        
        print()
        rand_ind = random.randint(0, len(valid_rows)-1)
        rand_row = valid_rows[rand_ind]
        rand_num_tallies = random.randint(1, tallies_left[rand_row])
        
        print("The computer took " + str(rand_num_tallies) + " tallies from the " + str(rand_row) + " row.")
        print("Updated Board:")
        
        for _ in range(rand_num_tallies):
            self.board[rand_row].pop()
        
        self.printWorld()
        return self.isOver()
        
    def player_move(self):
        """
        loops until valid user input for row and number of tallies to take and updates board 

        Parameters
        ----------
        none

        Returns
        -------
        self.is_over(): Bool of whether all rows are empty and game is over

        """
        valid_rows = []
        tallies_left = []
        for i in range(len(self.board)):
            if(map(len, self.board[i]) != 0):
                tallies_left.append(len(self.board[i]))
            if(len(self.board[i]) != 0):
                valid_rows.append(i)
        print()
        user_input = str (input("Enter which column {0,3} and number of tallies to take in comma separated list (col,#)"))

        row = user_input.split (",")[0]
        
        num_tallies = user_input.split (",")[1]
        if not row.isdigit() or not num_tallies.isdigit():
            print("Input as a comma separated list of numbers")
            self.player_move()
            return
        
        if not int(num_tallies) <= tallies_left[int(row)] or not int(row) in valid_rows:
            print("Invalid row or number of tallies. Try Again.")
            self.player_move()
            return
        
        print("You took " + str(num_tallies) + " tallies from the " + str(row) + " row.")
        print("Updated Board:")
        for _ in range(int(num_tallies)):
            self.board[int(row)].pop()
        
        self.printWorld()
        
        return self.isOver()

    def isOver(self):
        """
        checks to see if all rows of board are empty and if they are returns true

        Parameters
        ----------
        none

        Returns
        -------
        game_over(): Bool of whether all rows are empty and game is over

        """
        empty_rows = 0
        game_over = False
        for i in range(len(self.board)):
            if(len(self.board[i]) == 0):
                empty_rows += 1
        if(empty_rows == 4):
            game_over = True
        return game_over
    
```
