```python
'''
Author: Lauren Fisher
Class: CPSC323
Date: 10/21/2021
Desc: Main file handling game logic of Nim game of AI playing against itself until game is over
'''

from ipynb.fs.full.nim_board import Nim_Board
from ipynb.fs.full.nim_player import NimPlayer

NUM_PLAYERS = 2;

def playGame():
    """
    handles logic for game where developed ai plays against eachother following same strategy until board is empty

    Parameters
    ----------
    none

    Returns
    -------
    none

    """
    players = []

    for _ in range(NUM_PLAYERS):
        new_player = NimPlayer()
        players.append(new_player)

    nim_board = Nim_Board()
    nim_board.printWorld()
    print()
    
    while not nim_board.isEmpty():
        for player in players:
            # make a move here
            new_board = player.play(nim_board.state)
            nim_board.state = new_board
            nim_board.printWorld()
            print()
            if nim_board.isEmpty():
                break



def main():
    playGame()
```

```python
main()
```
